```python
#While loop
#Repeates the statements inside till the condition is True.When the condition becomes False,the loop stops
```

```python
#print first 3 numbers
count=-100
while(count<=3):
    print(count)
    count=count+1
```

```python
#print numbers in backward 3,2,1,0
count=3
while (count>=1):
    print(count)
    count-=1
```

```python
mystr="i got a job in big4"
#mystr.index('L')
i=0
#char='1'
indexes=[]
while i<len(mystr):
    if mystr[i]=='b':
        indexes.append(i)
    i=i+1
print("Indexes of l in the given string are:",indexes)
```

```python
#Control statements on loop
#break--will stop the loop
mystr='hello world'
i=0
while i<len(mystr):
    if mystr[i]==' ':
        break
    print(mystr[i])
    i=i+1
```

```python
#Control statements on loop
#break--will stop the loop
mystr='hello world'
i=0
while i<len(mystr):
    if mystr[i]==' ':
        break
    print(mystr[i],end=" ")
    i=i+1
```

```python
#continue statement -- is a keyword which stop the loop with the current iteration and contiues the loop with conditio checking for next iterations
i = 0
while i <= 5:
    i = i + 1
    if i == 3:
        continue
    print(i)
```

```python
i=1
while (i<=5):
    n=int(input("Enter a number:"))
    if n<0:
        print("Invalid choice")
        continue
    else:
        print(n)
    i=i+1
```

```python
#for loop-It is used to loop through the iterable elements like string,list.tuple etc.
list1=[12,3,7,8,9]
for i in list1:
    print(i*2)
```

```python
mystr="hello world"
for i in mystr:
    if i==" ":
        break
    print(i,end="  ")
```
