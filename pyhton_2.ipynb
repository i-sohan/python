##Comparision Operator

Compares one value with the another and return s the result in terms of true or false

1) > (greater than)
2) < (lesser than)
3) >= (greater than or equal to)
4) <= (Lesser than or equal to)
5) == (equal to)
6) != (not equal to)

```python
print(5>2)
print(5<2)
print(5>=2)
print(5<=2)
print(5==2)
print(5!=2)
```

##Logical Operators

They are used to check 2 or more comparisions
| A     | B     | A and B | A or B |
| ----- | ----- | ------- | ------ |
| True  | True  | True    | True   |
| True  | False | False   | True   |
| False | True  | False   | True   |
| False | False | False   | False  |


```python
#and --Logical operators--it returns True if both the statements are True otherwise it returns False
print(5>2 and 5<10)
print(5>2 and 5>10)
print(5<2 and 5<10)
print(5<2 and 5>10)
```

```python
print(5==5 and 6<8)
```

```python
#or --Logical operators--it returns True if at least one of the statements is True otherwise it returns False
print(5>2 or 5<10)
print(5>2 or 5>10)
print(5<2 or 5<10)
print(5<2 or 5>10)
```

```python
#not --Logical operators--it returns True if the statement is False otherwise it returns False
print(not(5>2))
```

```python
#Multiple logical operators then first it will check the and operator then it will check the or operator
print(5>2 and 5<10 or 5>10)
```
