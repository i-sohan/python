if statements

-if statements are conditional statements in python
-if condtion is true statements inside will be exceuted
-if condition is false staemts immediately after if will be executed
-indentation is the space where it specify the code

```python
condition = True  # You can change this to False to test the else block
if (condition):
    # Code to execute if the condition is True
    print("Condition is True")
else:
    # Code to execute if the condition is False
    print("Condition is False")
```

```python
x=5+1
if (x>=6):
    print("x is positive")
else:
    print("x is not positive")
```

```python
x='HELLO WORLD'
if (x.isupper()):
    print("x is in uppercase")
else:
    print("x is not in uppercase")
print("Bye")
```

#if-elif statements
-these are used to check the multiple conditions
-if one is not true another will be checked

```python
x=21
if x<0:
    print("first if block")
elif x==0:
    print("second if block")
elif x<10:
    print("third if block")
elif x>10:
    print("fourth if block")
elif x>20:
    print("fifth if block")
else:
    print("sixth if block")     
```

```python
#write a program using conditional statements based on age that you are eligible to get into cricket team and age criteria is greater than 18
age = int(input("Enter your age: "))
if age > 18 and age <= 50:
    print("You are eligible to get into the cricket team.")
elif age > 8 and age <= 18:
    print("You are not eligible to get into the cricket team.")
elif age >50:
    print("You are not eligible to get into the cricket team.")
else:
    print("no alphabets or special characters are allowed in age. Please enter a valid age.")
```

```python
#Nested if else statement - It is a conditional statement that allows you to execute different blocks of code based on multiple conditions.
#In order to execute inner if statement, the outer if statement must be true. 
# The basic syntax is:

if (condition1):
    # Code to execute if condition1 is True
    if (condition2):
        # Code to execute if condition2 is True
        print("Both conditions are True")
    else:
        # Code to execute if condition2 is False
        print("Condition1 is True but Condition2 is False")
```

```python
x=12
if x>0:
    print("First if block")
if x<15:
    print("Second if block")
    if x>10 and x<15:
        print("Third if block") 
```
