```python
#Python has 2 kinds of loops
#1. for loop
#2. while loop

```

```python
#The for Loop

#A for loop iterates over a sequence such as:
#String
#List
#Tuple
#Set
#Dictionary
#range()

#Syntax
for variable in sequence:
    # code
```

```python
for i in range(5):
    print(i)
```

```python
#Understanding the range() function
#1.range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and stops before a specified number.
```

```python
#range(start)-If the start parameter is specified, the sequence of numbers will start from that value.

for i in range(5):
      print(i)
```

```python
#range(start, stop)-If the start parameter is specified, the sequence of numbers will start from that value.

for i in range(2, 5):
      print(i)
```

```python
#range(start, stop, step)-If the step parameter is specified, the sequence of numbers will increment by that value (instead of 1).

for i in range(2, 10, 2):
      print(i)
```

```python
#Negative Indexing
#Negative indexing means beginning from the end, -1 refers to the last item, -2

for i in range(-1, -10, -1):
      print(i)
```

```python
#The while Loop-If a condition is true, the code inside the while loop is executed. After the code block has been executed, the condition is checked again. If it is still true, the code block is executed again. This continues until the condition becomes false.

#While loop syntax
while condition:
    # code
```

```python
#Example of while loop
count = 1

while count <= 5:
    print(count)
    count += 1
```

```python
#Break Statement-You can use the break statement to stop the loop even if the while condition is true:
while True:
    print("This will print forever.")
    break
```

```python
#continue Statement-You can use the continue statement to stop the current iteration, and continue with the next:
while True:
    print("This will print forever.")
    continue    
```

```python
#pass Statement-If you want to write a while loop with no content, put in the pass statement to avoid getting an error.
while True:
    pass
```

```python
#Nested Loops
#A nested loop is a loop inside a loop. The "inner loop" will be executed one time for each iteration of the "outer loop":
for i in range(3):
    for j in range(2):
        print(i, j)
```

```python
#loop with else statement-Loop statements may have an else block; it is executed when the loop is finished:
for i in range(5):
    print(i)
else:
    print("Loop finished.")

```
