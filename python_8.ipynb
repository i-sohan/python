```python
# FUNCTIONS
# A function is a block of reusable code that performs a specific task
# Syntax:
#   def function_name(parameters):
#       """docstring"""
#       statement(s)
#       return value
```

```python
# DOCSTRING
# A docstring is a string written inside a function to describe what it does
# Written using triple quotes """ """ right after the def line
# It helps others (and you!) understand the function later

# Single-line docstring
def addby10(x):
    """Adds 10 to the given number and returns the result."""
    return x + 10

print(addby10(5))        # 15
print(addby10.__doc__)   # prints the docstring
```

```python
# Multi-line docstring
# Used when the function needs more detailed explanation

def add(x, y):
    """
    Adds two numbers and returns their sum.

    Parameters:
        x (int/float): First number
        y (int/float): Second number

    Returns:
        int/float: Sum of x and y
    """
    return x + y

print(add(3, 5))     # 8
print(add.__doc__)   # prints the full docstring
```

```python
# help() -- another way to read a function's docstring

def greet(name):
    """Greets the user by their name."""
    print("Hello,", name)

help(greet)  # displays the docstring in a readable format
```

```python
# Example 1: Basic function (no parameters, no return)

def greet():
    """Prints a welcome message."""
    print("Hello! Welcome to Python Functions.")

greet()
```

```python
# Example 2: Function with a parameter

def greet_user(name):
    """Greets a specific user by name."""
    print("Hello,", name)

greet_user("Sohan")
greet_user("Alice")
```

```python
# Example 3: addby10 -- function that adds 10 to a number

def addby10(x):
    """Adds 10 to x and returns the result."""
    return x + 10

print(addby10(5))   # 15
print(addby10(20))  # 30
print(addby10(0))   # 10
```

```python
# Example 4: Function with multiple parameters

def add(x, y):
    """Returns the sum of x and y."""
    return x + y

print(add(3, 5))    # 8
print(add(10, 20))  # 30
```

```python
# Example 5: Function with default parameter value
# If no argument is passed, the default is used

def multiply(x, y=2):
    """Multiplies x by y. Default value of y is 2."""
    return x * y

print(multiply(5))     # 10  (uses default y=2)
print(multiply(5, 3))  # 15  (overrides default)
```

```python
# Example 6: Function returning multiple values

def min_max(numbers):
    """Returns the minimum and maximum values from a list."""
    return min(numbers), max(numbers)

low, high = min_max([3, 7, 1, 9, 4])
print("Min:", low)   # 1
print("Max:", high)  # 9
```

```python
# Example 7: Function calling another function

def square(x):
    """Returns the square of x."""
    return x * x

def sum_of_squares(a, b):
    """Returns the sum of squares of a and b."""
    return square(a) + square(b)

print(sum_of_squares(3, 4))  # 9 + 16 = 25
```

```python
# Example 8: Function with *args (variable number of arguments)
# *args lets you pass any number of arguments

def add_all(*numbers):
    """Returns the sum of any number of arguments passed."""
    total = 0
    for n in numbers:
        total += n
    return total

print(add_all(1, 2, 3))        # 6
print(add_all(10, 20, 30, 40)) # 100
```

```python
# Example 9: Lambda function (anonymous / one-line function)
# Syntax: lambda parameters: expression
# Note: Lambda functions do NOT support docstrings

addby10_lambda = lambda x: x + 10

print(addby10_lambda(5))   # 15
print(addby10_lambda(25))  # 35

# Lambda with two parameters
multiply = lambda x, y: x * y
print(multiply(4, 5))  # 20
```

```python
# Example 10: Using function with a list (apply addby10 to each element)

def addby10(x):
    """Adds 10 to x and returns the result."""
    return x + 10

numbers = [1, 2, 3, 4, 5]
result = list(map(addby10, numbers))
print(result)  # [11, 12, 13, 14, 15]
```

```python
# Example 11: Recursive function (function calling itself)
# Factorial: 5! = 5 * 4 * 3 * 2 * 1 = 120

def factorial(n):
    """
    Returns the factorial of a non-negative integer n.

    Parameters:
        n (int): A non-negative integer

    Returns:
        int: Factorial of n
    """
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # 120
print(factorial(6))  # 720
```
