```python
# 1. Given a string "Artificial Intelligence", extract:
# a) First 5 characters
# b) Last 5 characters
# c) Characters from index 3 to 10

text = "Artificial Intelligence"
print(text[:5])
print(text[-5:])
print(text[3:11])
```

```python
# 2. Replace "python" with "Java" in the string
text = "python programming"
print(text.replace("python", "Java"))
```

```python
# 3. Extract every second character from a string
text = "DataAnalytics"
print(text[::2])
```

```python
# 4. Remove the first and last characters using slicing
text = "MachineLearning"
print(text[1:-1])
```

```python
# 5. Check if a number lies between 10 and 50
num = int(input("Enter a number: "))
if 10 <= num <= 50:
    print("The number lies between 10 and 50")
else:
    print("The number does not lie between 10 and 50")
```

```python
# 6. Check whether a character is an alphabet, digit, or special character
ch = input("Enter a character: ")
if ch.isalpha():
    print("Alphabet")
elif ch.isdigit():
    print("Digit")
else:
    print("Special character")
```

```python
# 7. Validate a password
password = input("Enter password: ")

has_length = len(password) >= 8
has_digit = any(char.isdigit() for char in password)
has_upper = any(char.isupper() for char in password)

if has_length and has_digit and has_upper:
    print("Valid password")
else:
    print("Invalid password")
```

```python
# 8. List operations
list1 = [5, 12, 4, 6, 10]
list1.append(20)                      # a) Add to end
list1.insert(2, 7)                    # b) Insert at position 2
list1.remove(4)                       # c) Remove by value
list1.pop(1)                          # d) Remove by index
print(list1)
print(len(list1))                     # e) Length of list
```

```python
# 9. Tuple operations
t1 = (3, 6, 1, 7, 3, 9, 3)
print(t1.count(3))           # a) Count occurrences of 3
print(t1.index(9))          # b) Find index of 9
print(list(t1))             # c) Convert tuple to list
```

```python
# 10. Set operations
s1 = {5, 9, 10, 14, 15}
s2 = {9, 14, 18, 20, 25}
s1.remove(15)                 # a) Remove 15 from s1
print(s1)
print(s1 | s2)                # b) Union
print(s1 & s2)                # c) Intersection
print(s1 - s2)                # d) Difference
```

```python
# 11. Dictionary operations
d1 = {'name': 'xyz', 'age': 20, 'salary': 20000}
d1['city'] = 'Delhi'            # a) Add new key-value pair
d1['salary'] = 50000           # b) Update salary
d1.pop('age')                  # c) Remove age
# d) Create hobby with multiple values
d1['hobby'] = ['dancing', 'singing', 'painting']
print(d1)
```

```python
# 15. Function to find the minimum element in a list
def find_min(lst):
    return min(lst)

print(find_min([4, 7, 2, 9, 1]))
```

```python
# 16. Use reduce() with a lambda to find the product of list elements
from functools import reduce
list2 = [6, 2, 7, 1, 12]
product = reduce(lambda x, y: x * y, list2)
print(product)
```

```python
# 17. Create a function that applies different operations based on user input

def calculator(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    else:
        return "Invalid operator"

print(calculator(10, 5, '+'))
```
