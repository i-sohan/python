```python
#set-It is a unordered collection of elements put inside a {}
#dulicates are ignored in the set
#indexing is not possible in the set of elements
```

```python
s1=set()  #empty set
s1
```

```python
type(s1)
```

```python
s2={4,4,4,5,4,3,3,4,554,44,4}
s2
```

```python
s3={3,4,5,True,"hello",(2,4)}#set of mixed data
s3
```

```python
#set functions
```

```python
#add() -- which adds new vslue to the set
s2.add(5)
s2
```

```python
#update()--adds multiples value to the set
s2.update({8.5,-1.2,3.8,7})
s2
```

```python
#remove() --to remove based on the value
s2.remove(4)#It throws the error when element is not present in the set
s2
```

```python
#discard() -- removes based on the value,but it doesnt throw error if element is not present
s2.discard(-1.2)
s2
```

```python
#sorted() --sorts the set and returns list in the form of list
sorted(s2)
```

```python
#clear()--deletes all elements from set
s2.clear()
s2
```

```python
#Set operations
#union()-- returns all the elements from both the sets
#intersection() --returns common elements from the both the sets
#difference()-- returns elements present in one set but not present in other set
```

```python
a={1,2,3,4,5}
b={4,5,6,7}
c={7,8,9}
```

```python
a.union(b)
```

```python
b.union(a,c)
```

```python
a|b|c
```

```python
a.intersection(b)
```

```python
b.intersection(c)
```

```python
a&b
```

```python
a&b&c
```

```python
a.difference(b)
```

```python
b.difference(c)
```

```python
b-c
```

```python
type(d1)
```

```python
d2={5:'five',8:'eight',2:'two'} #dict with int keys
d2
```

```python
d3={'name':'Tom','age':20,'city':'goa'}#dict with string keys
d3
```

```python
marks={'science':90,'english':51,'math':70}
if marks.get('english',0) > 50: #get9key,default value)--retrieve value for a key
    print("PASS")
else:
    print("FAIL")
```
