```python
#Measure of Dispersion
#Range, Variance, Standard Deviation
```

```python
statistics.variance(ages)
```

```python
np.var(ages,axis=0)
```

```python
def variance(data):
    n = len(data)
    mean = sum(data) / n
    deviations = [(x - mean) ** 2 for x in data]
    variance = sum(deviations) / n
    return variance
```

```python
variance(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

```python
#log normal distribution,power law distribution
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.lognormal(mu, sigma, 1000)
```

```python
sns.histplot(s, kde=True)
```

```python
sns.histplot(np.log(s), kde=True)
```

```python
#check whether the data is normally distributed or not
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab
def plot_data(sample):
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    sns.histplot(sample, kde=True)
    plt.subplot(1, 2, 2)
    stats.probplot(sample, dist="norm", plot=pylab)
    plt.show()
```

```python
s=np.random.normal(loc=0, scale=1, size=1000)
plot_data(s)
```
