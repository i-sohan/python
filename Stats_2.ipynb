```python
#collect the data
old_scores = [65, 70, 68, 72, 66, 69, 71, 67]
new_scores = [75, 78, 80, 77, 79, 76, 81, 78]
```

```python
import pandas as pd

df = pd.DataFrame({
    "Old": old_scores,
    "New": new_scores
})

print(df)
```

```python
#calculate the mean and standard deviation of the old and new scores
print(df.mean())
```

```python
#State the hypotheses
#Null Hypothesis
#Average scores are the same.

#Alternative Hypothesis
#Average scores are the same.
```

```python
#Perform Independent T-test
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(old_scores, new_scores)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
#Interpret
alpha = 0.05

if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
from statsmodels.stats.proportion import proportions_ztest

success = [120,150]

total = [1000,1000]

z_stat,p = proportions_ztest(success,total)

print(z_stat)

print(p)
```

```python
#Decision value
0.039 < 0.05
Conclusion

The new webpage significantly increased conversions.

This exact test is used in A/B testing for websites.
```

```python
#Before vs After (Paired T-Test)
before = [150,145,160,155,148]
```

```python
after = [140,138,150,147,142]
```

```python
from scipy.stats import ttest_rel

t,p = ttest_rel(before,after)

print(t)

print(p)
```

```python
#Compare Three Groups (ANOVA)
from scipy.stats import f_oneway

A=[65,67,66,64,68]

B=[75,76,77,74,78]

C=[82,80,83,81,84]

f,p=f_oneway(A,B,C)

print(f)

print(p)
```

```python
#if 
p<0.05
```
